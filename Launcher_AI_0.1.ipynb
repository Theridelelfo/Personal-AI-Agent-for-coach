```python
# === COLAB CELL 1: PACK (pulizia totale) ===
# Cancella tutto /content e copia TUTTO runtime/* direttamente sotto /content/

from pathlib import Path
import shutil, os

REPO_URL   = "https://github.com/Theridelelfo/Personal-AI-Agent-for-coach.git"  # sostituisci <user>
REPO_LOCAL = Path("/content/repo")   # cache del clone del repo
CONTENT    = Path("/content")

# 1) Pulizia totale di /content (eccetto sample_data opzionale)
for item in CONTENT.iterdir():
    shutil.rmtree(item) if item.is_dir() else item.unlink()

# 2) Clone o pull del repository
if REPO_LOCAL.exists():
    print("[PACK] Repo già presente → git pull")
    os.system(f"cd {REPO_LOCAL} && git pull")
else:
    print("[PACK] Clono il repo...")
    os.system(f"git clone {REPO_URL} {REPO_LOCAL}")

# 3) Copia TUTTO il contenuto di runtime/ → /content/
src = REPO_LOCAL / "Runtime"
assert src.exists(), "Cartella Runtime mancante nel repo"

for item in src.iterdir():
    target = CONTENT / item.name
    if item.is_dir():
        shutil.copytree(item, target)
    else:
        shutil.copy2(item, target)

print("[PACK] Copia completata: runtime/* → /content/")
print("[PACK] Pronto per SETUP (cella 2).")
```
